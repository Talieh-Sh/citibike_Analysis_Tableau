```python
import pandas as pd
```

```python
# File paths
file1 = 'Resources/JC-202312-citibike-tripdata.csv'
file2 = 'Resources/JC-202311-citibike-tripdata.csv'
file3 = 'Resources/JC-202310-citibike-tripdata.csv'

# Read the files
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)

# Concatenate the dataframes
combined_df = pd.concat([df1, df2, df3])

# Check the combined dataframe
combined_df.head()
```

```python
combined_df['rideable_type'].unique()
```

```python
combined_df.info()
```

```python
# Count the number of missing values in each column
missing_values_count = combined_df.isnull().sum()
missing_values_count
```

```python
#Drop rows with any missing values
combined_df = combined_df.dropna()
missing_values_count = combined_df.isnull().sum()

missing_values_count
```

```python
#Convert Data Types
combined_df['started_at'] = pd.to_datetime(combined_df['started_at'])
combined_df['ended_at'] = pd.to_datetime(combined_df['ended_at'])
```

```python


# Extract the Month component and create a new column
combined_df['start_month'] = combined_df['started_at'].dt.month


# Extract the Month component and create a new column
combined_df['end_month'] = combined_df['ended_at'].dt.month
```

```python
combined_df.head()
```

```python

# Comparing two columns, start_month and end_month
equal_data = combined_df['start_month'] == combined_df['end_month']

# Check the result
print(equal_data)

# To count how many rows have equal data
equal_count = equal_data.sum()
print(f"Number of rows with equal data in 'column1' and 'column2': {equal_count}")

# To check if all values are equal
all_equal = equal_data.all()
print(f"All values are equal: {all_equal}")
```

from above it shows not all start months are equal to end months we consider the month as start month

```python
# Create a mapping from month number to month name
month_mapping = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
    7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Use map to create the new column
combined_df['month'] = combined_df['start_month'].map(month_mapping)

# Check the result
combined_df['month'].unique()
```

```python
combined_df.head()
```

```python
# Save the cleaned data
combined_df.to_csv('Resources/cleaned_citibike_tripdata.csv', index=False)
```
